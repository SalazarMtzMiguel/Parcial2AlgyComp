```python
class NodoArbol:
    def __init__(self, valor):
        self.value = valor
        self.izq = None
        self.Der = None


def armaArbol(preorder, inorder):
    if not preorder or not inorder:
        return None

    # La raiz es el primer elemento del preorden
    valor_raiz = preorder[0]
    raiz = NodoArbol(valor_raiz)

    # Encuentra el indice de la raiz en el recorrido en orden
    posicionRaiz = len(inorder) - 1 - inorder[::-1].index(valor_raiz)

    # Divide en subarboles izquierdo y derecho
    inorder_izq = inorder[:posicionRaiz]
    inorder_der = inorder[posicionRaiz + 1:]

    preorder_izq = preorder[1:1 + len(inorder_izq)]
    preorder_der = preorder[1 + len(inorder_izq):]

    # Construye recursivamente los subarboles
    raiz.izq = armaArbol(preorder_izq, inorder_izq)
    raiz.Der = armaArbol(preorder_der, inorder_der)

    return raiz


# Funcion para imprimir el arbol (por ejemplo, recorrido en orden)
def print_inorder(node):
    if node:
        print_inorder(node.izq)
        print(node.value, end=" ")
        print_inorder(node.Der)
# Funcion para calcular profundidad (luego se usa para imprimir el arbol)
def calculate_depth(node):
    if not node:
        return 0
    left_depth = calculate_depth(node.izq)
    right_depth = calculate_depth(node.Der)
    return max(left_depth, right_depth) + 1

# Funcion para imprimir el arbol de forma mas visual
def print_tree(node, level=0, prefix="Raiz: "):
    if node is not None:
        print(" " * (level * 4) + prefix + str(node.value))
        if node.izq is not None or node.Der is not None:
            if node.Der:
                print_tree(node.Der, level + 1, "D--- ")
            else:
                print(" " * ((level + 1) * 4) + "D--- None")
            if node.izq:
                print_tree(node.izq, level + 1, "I--- ")
            else:
                print(" " * ((level + 1) * 4) + "I--- None")



# Ejemplo de uso
preorder = [10, 5, 2, 7, 15, 12, 20]
inorder = [2, 5, 7, 10, 12, 15, 20]

tree = armaArbol(preorder, inorder)
print("Recorrido en orden del arbol reconstruido:")
print_tree(tree)


# Ejemplo de uso
preorder = [30,20,10,15,25,40,35,50]
inorder = [10,15,20,25,30,35,40,50]

tree = armaArbol(preorder, inorder)
print("Recorrido en orden del arbol reconstruido:")
print_tree(tree)

# Ejemplo de uso
preorder = [8,3,3,6,4,7,10,9,14,13]
inorder = [3,3,4,6,7,8,9,10,13,14]

tree = armaArbol(preorder, inorder)
print("Recorrido en orden del arbol reconstruido:")
print_tree(tree)
```
